```python
df = pd.read_csv('Unemployment in India.csv')  # Change file name if different
df.head()
```

```python
# Check data types and non-null counts
df.info()

# Check for missing values in each column
df.isnull().sum()
```

```python
# Remove any rows with missing values (if there are any)
df.dropna(inplace=True)
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()  # Removes leading/trailing spaces

# Now you can safely convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Convert the 'Date' or 'Month' column to datetime (adjust the column name as per your dataset)
df['Date'] = pd.to_datetime(df['Date'])  # Replace 'Date' if your column name is different
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a new column to label data as Pre-COVID or Post-COVID
df['Period'] = df['Date'].apply(lambda x: 'Pre-Covid' if x < pd.to_datetime('2020-03-01') else 'Post-Covid')
```

```python
# Create a new column to label data as Pre-COVID or Post-COVID
df['Period'] = df['Date'].apply(lambda x: 'Pre-Covid' if x < pd.to_datetime('2020-03-01') else 'Post-Covid')

# Correct box plot with actual column name
plt.figure(figsize=(8,5))
sns.boxplot(x='Period', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Before vs After COVID-19')
plt.xlabel('Period')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
# Extract month from the date
df['Month'] = df['Date'].dt.month

# Plot monthly trend using correct column name
plt.figure(figsize=(10, 5))
sns.lineplot(x='Month', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Seasonal Unemployment Trend')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```
